```python
class Matrix:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def addition(self): #addition of two matrices 
        return self.a+self.b
    def elesubstraction(self): #substraction
        return self.a-self.b
    def elemultiplication(self): #element wise multiplication
        return self.a*self.b
    def scalarmulti(self,scalar):  #scalar multiplication of matrices 
        return  scalar*self.a,scalar*self.b
    def determinant(self):  #fdeterminant of both matrices 
        return np.linalg.det(self.a),np.linalg.det(self.b)
    def marix_mul(self):  #Matrix multiplication
        return self.a@self.b
    def Transpose(self): #transpose of both matrices  
        return self.a.T,self.b.T
    def inverse(self): #inverse of both matrices 
        return np.linalg.inv(self.a),np.linalg.inv(self.b)
    def diagonal(self): #diagonal elements of first matrix
        print(diag(self.a))
    def rank(self):  #rank of both matrix
        return np.linalg.matrix_rank(self.a),np.linalg.matrix_rank(self.b)
    def eigen(self):  #eigenvalues and eigenvectors of first matrix.
        a=np.linalg.eig(self.a)[0]  #eigenvalues 
        b=np.linalg.eig(self.a)[1]  #eigenvectors 
        if len(self.a)==2:
            x=np.linspace(0,b[0][0],100)
            y=np.linspace(0,b[1][0],100)
            ax=plt.plot(x,y)
            x=np.linspace(0,b[0][1],100)
            y=np.linspace(0,b[1][1],100)
            ax=plt.plot(x,y)
            plt.show()
        elif len(self.a)==3:
            t=np.linspace(0,1,100)
            x=t*b[0][0]
            y=t*b[1][0]
            z=t*b[2][0]
            fig=plt.figure()
            ax=fig.add_subplot(111,projection='3d')
            ax.plot(x,y,z)
            t=np.linspace(0,1,100)
            x=t*b[0][1]
            y=t*b[1][1]
            z=t*b[2][1]
            #fig=plt.figure()
            #ax=fig.add_subplot(111,projection='3d')
            ax.plot(x,y,z)
            t=np.linspace(0,1,100)
            x=t*b[0][2]
            y=t*b[1][2]
            z=t*b[2][2]
            #fig=plt.figure()
            #ax=fig.add_subplot(111,projection='3d')
            ax.plot(x,y,z)
            plt.show()
        else:
            print('We can not plot the eigenvectors')
        return a,b
m=Matrix(np.array([[1,2,3],[2,3,4],[3,4,5]]),np.array([[1,2,3],[2,3,4],[3,4,5]]))
m.eigen()
m1=Matrix(np.array([[1,2],[3,4]]),np.array([[3,4],[5,6]]))
m1.eigen()
```

```python
#(4) z=sin(sqrt(x^2+y^2))
def f(x):
    return np.sin(np.sqrt(x[0]**2+x[1]**2))
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
xx,yy=np.meshgrid(x,y)
zz=f(np.array([xx,yy]))
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(xx,yy,zz,color='g')
ax.set(xlabel='x',ylabel='y',title='f(x)',zlabel='z')
plt.show()
```

```python
#first i am definig all the sigmoid function
def f1(x):
    return 1/(1+np.exp(-(x[0]+x[1]+1)))
def f2(x):
    return 1/(1+np.exp(-(x[0]+x[1]+x[0]*x[1])))
def f3(x):
    return 1/(1+np.exp(-(x[0]**2+x[1]**2+x[0]*x[1])))
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
xx,yy=np.meshgrid(x,y)
zz=f1(np.array([xx,yy]))
plt.contour(xx,yy,zz)
plt.show()
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
xx,yy=np.meshgrid(x,y)
zz=f2(np.array([xx,yy])) 
plt.contour(xx,yy,zz)
plt.show()
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
xx,yy=np.meshgrid(x,y)
zz=f3(np.array([xx,yy]))
cp=plt.contour(xx,yy,zz)
plt.clabel(cp,fontsize=5)
plt.show()
```
