# 2.(a)

```python
# Import the necessary libraries
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# Load the Train dataset
train_data=pd.read_csv(r'C:\\Users\\MANOJ PATEL\\Downloads\\Train_dataset.csv')
#reshaping the data
X_train =np.array(train_data['x']).reshape(-1,1)
y_train = np.array(train_data['y']).reshape(-1,1)
#Loading the test Data
test_data=pd.read_csv(r'C:\\Users\\MANOJ PATEL\\Downloads\\Test_dataset.csv')
#Again reshaping
X_test=np.array(test_data['x']).reshape(-1,1)
y_test=np.array(test_data['y']).reshape(-1,1)

# Transform the training data into a polynomial of degree 3
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X_train)

# Fit a linear regression model to the transformed training data
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y_train)
# Make predictions on the test dataset
X_test_poly = poly_reg.transform(X_test)
y_pred = lin_reg.predict(X_test_poly)

# Compute the mean square error of the predictions on the test dataset
mse = mean_squared_error(y_test, y_pred)
print("MSE without regularization: ", mse)

# Print the final weights of the model
print("Final weights without regularization: ", lin_reg.coef_)
#Here are the final results
results=pd.DataFrame({'Actual':y_test.flatten(),'Predicted':y_pred.flatten()})
```

# 2.(b)

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

train_data=pd.read_csv(r'C:\\Users\\MANOJ PATEL\\Downloads\\Train_dataset.csv')
X_train =np.array(train_data['x']).reshape(-1,1)
y_train = np.array(train_data['y']).reshape(-1,1)
X_test=np.array(test_data['x']).reshape(-1,1)
y_test=np.array(test_data['y']).reshape(-1,1)
# Normalize the training and test sets
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Perform polynomial regression of degree 3 with L2 regularization
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
ridge = Ridge(alpha=0.001)  #alpha is Regularization Parameter  if alpha=0 it is usual above one.
ridge.fit(X_train_poly, y_train)

# Predict the target variable for the test set
y_pred = ridge.predict(X_test_poly)

# Calculate the mean square error for the test set
mse = mean_squared_error(y_test, y_pred)
print('Mean Square Error :', mse)

# Print the final weights of the model
weights = np.append(ridge.intercept_, ridge.coef_)
print('Final Weights with regularization:', weights)
```

```python
results=pd.DataFrame({'Actual':y_test.flatten(),'Predicted':y_pred.flatten()})
results
```
